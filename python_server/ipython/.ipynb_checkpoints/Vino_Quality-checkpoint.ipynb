```python
X_std = scale(X)
```

```python
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
estimators = {'k_means_wine_2': KMeans(n_clusters=2),
              'k_means_wine_3': KMeans(n_clusters=3),
              'k_means_wine_8': KMeans(n_clusters=8),
              'k_means_wine_bad_init': KMeans(n_clusters=3, n_init=1,
                                              init='random')}
```

```python
fignum = 1
for name, est in estimators.items():
    fig = plt.figure(fignum, figsize=(8, 8))
    plt.clf()
    #ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=38, azim=134)
    #ax = fig.add_subplot(111, projection='3d')
    plt.cla()
    est.fit(X_std)
    labels = est.labels_
    centers = est.cluster_centers_
    
    """"
    #Tomado de http://www.gadatascience.com/modeling/kmeans.html 
    clusters = est.predict(X_std)
    results = pnd.DataFrame({ 'cluster' : clusters, 'kind' : X['kind']})
    cluster_counts = results.groupby('cluster')['kind'].value_counts()
    for i in xrange(len(cluster_counts)):
        print("Cluster " ,i)
        print(cluster_counts[i])
    """"
    #print centers
    ax.scatter(X_std[:, 3], X_std[:, 0], X_std[:, 2], c=labels.astype(np.float))
    ax.scatter(centers[:, 3], centers[:, 0], centers[:, 2], marker='x', s=350, linewidths=3, color='black')

    ax.set_title(name)
    #ax.w_xaxis.set_ticklabels([])
    #ax.w_yaxis.set_ticklabels([])
    #ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel('chlorides')
    ax.set_ylabel('fixed acidity')
    ax.set_zlabel('citric acid')
    fignum = fignum + 1
```
